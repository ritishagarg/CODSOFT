```python
import random
from tensorflow.keras.models import load_model
import json
import pickle
import numpy as np
import nltk
from nltk.stem import WordNetLemmatizer
from colorama import Fore


lemmatizer = WordNetLemmatizer()
replies = json.loads(open('reps.json').read())
words = pickle.load(open('words.pkl', 'rb'))
classes = pickle.load(open('classes.pkl', 'rb'))
model = load_model('chatbotmodel.h5')

def d1(sentence):
    wsentence = nltk.word_tokenize(sentence)
    wsentence = [lemmatizer.lemmatize(word)  for word in wsentence]
    return wsentence

def d2(sentence):
    wsentence= d1(sentence)
    bag = [0] * len(words)
    for w in wsentence:
        for i, word in enumerate(words):
            if word == w:
                bag[i] = 1

    return np.array(bag)

def d3(sentence):
    bow = d2(sentence)
    res = model.predict(np.array([bow]))[0]
    error = 0.25
    results = [[i,r] for i, r in enumerate(res) if r > error]

    results.sort(key=lambda  x:x[1], reverse=True)
    return_list = []
    for r in results:
        return_list.append({'intent': classes[r[0]], 'probability': str(r[1])})
    return return_list


def resp(replies_list,replies_json):
    tag= replies_list[0]['intent']
    list_of_intents =replies_json['intents']
    
    for i in list_of_intents:
        if i['tag'] == tag:
            result = random.choice(i['responses'])
            break
    return result

print(Fore.RED + "       WELCOME TO THE CHATBOT          ")

while True:
    show_message = input(Fore.BLUE + "| You: ")
    if show_message == "bye" or show_message == "Goodbye":
        ints = d3(show_message)
        res = resp(ints, replies)
        print("| ChatBot:", res)
        print(Fore.RED + "|         Program Ended         |")
        exit()

    else:
        ints = d3(show_message)
        res = resp(ints, replies)
        print(Fore.GREEN + "| ChatBot:", res)
```
